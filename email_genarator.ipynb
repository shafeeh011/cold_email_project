```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

load_dotenv()
groq_api_key = os.getenv("GROQ_API_KEY")
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    max_tokens=None,
    groq_api_key=groq_api_key
)
```

```python
resposnse = llm.invoke(
    "Generate a cold email template"
)

print(resposnse.content)
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(
    "https://careers.nike.com/analyst-finance-cost-and-standards-engineering/job/R-52822"
)

page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}

    ### EXTRACT THE FOLLOWING INFORMATION:
    The Scraped text from the career's page of website.
    Your job is to extract the job posting and return them in JSON format containing the following keys: 'title', 'skills', 'company', 'location', 'description', 'experience'.
    Only return the JSON format, nothing else.
    ### VALID JSON FORMAT(NO PREAMBLE):
    """
)
chain_extract = prompt_extract | llm
response1 = chain_extract.invoke(input={'page_data': page_data})
print(response1.content)
```

```python
import json
response1_json = json.loads(response1.content)
print(response1_json)
type(response1_json)
```

```python
import pandas as pd

df = pd.read_csv("/home/muhammed-shafeeh/AI_ML/cold_email_project/my_portfolio.csv")
```

```python
import chromadb
import uuid


chroma_client = chromadb.PersistentClient('vectorstore')
collection = chroma_client.get_or_create_collection(name="my_portfolio_collection")

for _, row in df.iterrows():
    collection.add(
        documents=row["Techstack"],
        metadatas={"link": row["Links"]},
        ids=[str(uuid.uuid4())],
    )
```

```python
collection.get()
```

```python
links = collection.query(
    query_texts=["response1_json"],
    n_results=1
)
print(links)
links = links.get("metadatas")
```

```python
job = response1_json
job["skill"] 
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}
        
        ### INSTRUCTION:
        You are Mohan, a business development executive at AtliQ. AtliQ is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools. 
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability, 
        process optimization, cost reduction, and heightened overall efficiency. 
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of AtliQ 
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Atliq's portfolio: {link_list}
        Remember you are Mohan, BDE at AtliQ. 
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
        Pleas provide.
        ### THE GREETING SHOULD BE "Dear Hiring Manager"
        """
        )

chain_email = prompt_email | llm
response2 = chain_email.invoke(input={
    'job_description': str(job),
    'link_list': links
})
print(response2.content)
```
